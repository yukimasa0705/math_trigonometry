# 三角比の基礎

## 1. 三角比とは？

三角比とは、**直角三角形**において、角度と辺の長さの関係を表す比率のことや！  
特に「サイン（sin）」「コサイン（cos）」「タンジェント（tan）」の3つが基本やで。

ある直角三角形において、角 $\theta$ に対して以下のように定義される。

- $\sin\theta = \dfrac{\text{向かい側の辺の長さ}}{\text{斜辺の長さ}}$
- $\cos\theta = \dfrac{\text{隣の辺の長さ}}{\text{斜辺の長さ}}$
- $\tan\theta = \dfrac{\text{向かい側の辺の長さ}}{\text{隣の辺の長さ}}$

---

## 2. 直角三角形と三角比

例えばこんな三角形を考えてみるで！

- 斜辺の長さ：$5$
- 向かい側の辺の長さ：$3$
- 隣の辺の長さ：$4$

このとき、角 $\theta$ に対する三角比はこうなる。

- $\sin\theta = \dfrac{3}{5}$
- $\cos\theta = \dfrac{4}{5}$
- $\tan\theta = \dfrac{3}{4}$

---

## 3. 代表的な角度の三角比

よく出てくる角度の三角比は、パターン覚えた方が早いで！

| 角度 | $\sin\theta$ | $\cos\theta$ | $\tan\theta$ |
|:----:|:------------:|:------------:|:------------:|
| $30^\circ$ | $\dfrac{1}{2}$ | $\dfrac{\sqrt{3}}{2}$ | $\dfrac{1}{\sqrt{3}}$ |
| $45^\circ$ | $\dfrac{\sqrt{2}}{2}$ | $\dfrac{\sqrt{2}}{2}$ | $1$ |
| $60^\circ$ | $\dfrac{\sqrt{3}}{2}$ | $\dfrac{1}{2}$ | $\sqrt{3}$ |

これらは、試験でも実戦でもめっちゃ使うから、絶対覚えとこな！

---

## まとめ

- 三角比は直角三角形における「辺の比」のこと！
- $\sin, \cos, \tan$ の意味をしっかり覚えよう。
- $30^\circ, 45^\circ, 60^\circ$ の三角比は即答できるようにしとこな！

---


# 単位円と三角関数

## 1. 単位円とは？

単位円っていうのは、**中心が原点 $(0,0)$、半径が $1$ の円**のことや！

単位円の式は、こうなるで。

$$
x^2 + y^2 = 1
$$

この単位円を使うと、三角比を角度だけで考えられるようになるんや！

---

## 2. 単位円と三角比の関係

単位円上の点 $P(x, y)$ を考えるで。  
ここで角度 $\theta$ は、正の向きに（反時計回りに）回した角度や。

このとき、三角比はこんなふうに表される！

- $\cos\theta = x$
- $\sin\theta = y$
- $\tan\theta = \dfrac{y}{x}$（ただし $x \neq 0$）

つまり、単位円を使えば、角度 $\theta$ から直接 $\sin, \cos, \tan$ を読み取れるってわけやな！

---

## 3. 各象限での符号

単位円では、角度によって $\sin, \cos, \tan$ の**符号**（プラスかマイナスか）が変わるねん。

| 象限 | $\sin\theta$ | $\cos\theta$ | $\tan\theta$ |
|:----:|:------------:|:------------:|:------------:|
| 第1象限 $(0^\circ～90^\circ)$ | ＋ | ＋ | ＋ |
| 第2象限 $(90^\circ～180^\circ)$ | ＋ | － | － |
| 第3象限 $(180^\circ～270^\circ)$ | － | － | ＋ |
| 第4象限 $(270^\circ～360^\circ)$ | － | ＋ | － |

この符号パターン、めっちゃ大事やから絶対押さえとこな！

---

## 4. 具体例

例えば、$\theta = 150^\circ$ のとき。

- 第2象限やから、$\sin\theta$ は正、$\cos\theta$ は負になる。
- $\theta = 150^\circ$ は $180^\circ - 30^\circ$ やから、基準角は $30^\circ$。
- よって、
  - $\sin 150^\circ = \sin 30^\circ = \dfrac{1}{2}$
  - $\cos 150^\circ = -\cos 30^\circ = -\dfrac{\sqrt{3}}{2}$

---

## まとめ

- 単位円は、半径1の円：$x^2 + y^2 = 1$
- $\sin\theta$ は $y$座標、$\cos\theta$ は $x$座標
- $\tan\theta = \dfrac{y}{x}$（ただし $x\neq0$）
- 象限ごとの符号パターンを覚えよう！

---


# 三角関数のグラフ

## 1. $\sin\theta$ のグラフ

まずは $\sin\theta$ のグラフからや！

- 形：なみなみ（波形）
- 周期：$360^\circ$（または $2\pi$）
- 振幅（上下の大きさ）：$1$
- 基本形：

$$
y = \sin\theta
$$

### ポイント
- $\theta = 0^\circ$ で $y=0$
- $\theta = 90^\circ$ で $y=1$
- $\theta = 180^\circ$ で $y=0$
- $\theta = 270^\circ$ で $y=-1$
- $\theta = 360^\circ$ で $y=0$

こういう流れで波を1回描くんや！

---

## 2. $\cos\theta$ のグラフ

次に $\cos\theta$ のグラフ！

- 形：これもなみなみ（波形）
- 周期：$360^\circ$（または $2\pi$）
- 振幅：$1$
- 基本形：

$$
y = \cos\theta
$$

### ポイント
- $\theta = 0^\circ$ で $y=1$
- $\theta = 90^\circ$ で $y=0$
- $\theta = 180^\circ$ で $y=-1$
- $\theta = 270^\circ$ で $y=0$
- $\theta = 360^\circ$ で $y=1$

$\sin\theta$ のグラフを $\frac{90^\circ}{4分の1周期}$ 左にずらした感じやな！

---

## 3. $\tan\theta$ のグラフ

最後に $\tan\theta$ のグラフ！

- 形：上下にビューンと無限に伸びる感じ
- 周期：$180^\circ$（または $\pi$）
- 振幅：なし（無限にいくから）
- 基本形：

$$
y = \tan\theta
$$

### ポイント
- $\theta = 0^\circ$ で $y=0$
- $\theta = 90^\circ$ で定義されない（**鉛直のアシンポート（漸近線）**が立つ）
- $\theta = 180^\circ$ でまた $y=0$
- $\theta = 270^\circ$ でまた定義されない
- $\theta = 360^\circ$ でまた $y=0$

$\tan$ は $\cos\theta = 0$ になるところでグラフがぶち切れる（漸近線になる）んやで！

---

## まとめ

| 関数 | 周期 | 振幅 | 特徴 |
|:----:|:----:|:----:|:-----:|
| $\sin\theta$ | $360^\circ$ | あり（$1$） | なみなみ |
| $\cos\theta$ | $360^\circ$ | あり（$1$） | なみなみ |
| $\tan\theta$ | $180^\circ$ | なし | 無限に伸びる |

---

## イメージ（簡単に）

- $\sin$：0から上がって、下がって、また戻る
- $\cos$：1から下がって、上がって、また戻る
- $\tan$：ビューンって無限に伸びる→リセット→またビューン！

---


```python
# ライブラリインポート
import numpy as np
import matplotlib.pyplot as plt

# 角度θを 0°～360°（ラジアンに直して）作成
theta = np.linspace(0, 2 * np.pi, 1000)

# sinθ, cosθ, tanθの値
sin_theta = np.sin(theta)
cos_theta = np.cos(theta)
tan_theta = np.tan(theta)

# 図を3つ並べて描画
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# --- sinθのグラフ ---
axs[0].plot(np.degrees(theta), sin_theta, label=r'$y=\sin\theta$', color='blue')
axs[0].set_title(r'$\sin\theta$のグラフ')
axs[0].set_xlabel('角度 (°)')
axs[0].set_ylabel('y')
axs[0].grid(True)
axs[0].legend()
axs[0].set_xticks(np.arange(0, 361, 90))
axs[0].set_yticks(np.arange(-1, 1.1, 0.5))
axs[0].set_xlim(0, 360)
axs[0].set_ylim(-1.2, 1.2)

# --- cosθのグラフ ---
axs[1].plot(np.degrees(theta), cos_theta, label=r'$y=\cos\theta$', color='green')
axs[1].set_title(r'$\cos\theta$のグラフ')
axs[1].set_xlabel('角度 (°)')
axs[1].set_ylabel('y')
axs[1].grid(True)
axs[1].legend()
axs[1].set_xticks(np.arange(0, 361, 90))
axs[1].set_yticks(np.arange(-1, 1.1, 0.5))
axs[1].set_xlim(0, 360)
axs[1].set_ylim(-1.2, 1.2)

# --- tanθのグラフ ---
# tanは値が発散するところを考慮して範囲制限
axs[2].plot(np.degrees(theta), tan_theta, label=r'$y=\tan\theta$', color='red')
axs[2].set_title(r'$\tan\theta$のグラフ')
axs[2].set_xlabel('角度 (°)')
axs[2].set_ylabel('y')
axs[2].grid(True)
axs[2].legend()
axs[2].set_xticks(np.arange(0, 361, 90))
axs[2].set_xlim(0, 360)
axs[2].set_ylim(-5, 5)  # 上下5くらいに制限しとく
# 90°, 270°に破線で漸近線を追加
for angle in [90, 270]:
    axs[2].axvline(x=angle, color='gray', linestyle='--')

# グラフ全体のレイアウト調整
plt.tight_layout()
plt.show()
```
